```python
my_list = [1,2,3,4,5] # creating original list
print(my_list)

my_slice = my_list[0:4] # slicing the list and assigning to a new list

my_slice[1] = 99 # changing my_slice values

print(my_slice)

print(my_list)

# You can see that my_list remains unchanged even though my_slice was changed
```

```python
mona_slice = monalisa[0:100, 0:100, :] # slicing Monalisa numpy array

mona_slice[:, :, :] = 155 # changing 'monaslice' values

pp.imshow(monalisa) # see the top-left box in gray. This shows that original Numpy array was modified

# Hence always use .copy() for creating copies of Numpy arrays
```

```python
monalisa= np.load('C:/Users/Public.DESKTOP-6RBQT7L/Desktop/Programming - Maths/LinkedIn Learning - Data analysis Python course/chapter4/monalisa.npy')

mona_slice2 = monalisa.copy()

mona_slice2[:100, :200, :] = 198

pp.imshow(mona_slice2) # Change in Mona_slice2 as there is a gray box
```
